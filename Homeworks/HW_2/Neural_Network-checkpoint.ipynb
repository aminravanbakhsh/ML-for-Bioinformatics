```python
#Dependencies
import numpy as np
import pandas as pd
import sklearn
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
import keras
from keras.models import Sequential
from keras.layers import Dense
import tensorflow as tf
```

```python
'''
    1 - import data from colab
    2 - drop unnamed
'''

d_path = '/content/drive/MyDrive/ML_for_BIO/hws/hw2/'
train_df = pd.read_csv(d_path + 'heart_train.csv')
test_df = pd.read_csv(d_path + 'heart_test.csv')

train_df = train_df.drop(columns = ['Unnamed: 0'])
test_df = test_df.drop(columns = ['Unnamed: 0'])
```

```python
'''
    split target and features
'''
train_y = train_df['target']
test_y = test_df['target']
train_X = train_df.drop(columns=['target'])
test_X = test_df.drop(columns=['target'])
```

<div style="direction:rtl;line-height:300%;text-align:justify;" align="justify"><font face="B nazanin" size=4>
تابع زیر دیتافریم می‌گیرد , پس از انجام پیش‌پردازش‌های لازم و تقسیم‌بندی به دو دسته,داده‌های آموزش و اعتبارسنجی را برمی‌گرداند
</font></div>

```python
def preprocessing(X, y, test_rate):
    '''
        return X_train, X_val, y_train, y_val
    '''
    return sklearn.model_selection.train_test_split(X, y,)
```

```python
X_train, X_val, y_train, y_val = preprocessing(train_X, train_y, 0.2)
```

<div style="direction:rtl;line-height:300%;text-align:justify;" align="justify"><font face="B nazanin" size=4>
تابع زیر داده‌ی آموزش و هایپرپارامتر‌های موردنیاز را (که به صورت دیکشنری است و یکی از کلید‌های آن layers بوده که مقدار آن  آرایه‌ای است شامل پارامتر‌های مربوط به هر لایه)را می‌گیرد
    و پس از پیاده‌سازی مدل و آموزش آن, مدل ترین‌شده را بازمی‌گرداند
</font></div>

```python
def train_model(X_train, y_train, hidden_layers_number, units_array, activations_array, loss_function, metrics_array, learning_rate, batch_size, epochs, verbose):
    '''
        X_train: pd data frame
        y_tarin: pd series

        hparams:
            hidden layers number:
            array of units
            arraya of activations
            loss function
            metrics array
            learning rate
            batch_size
            epochs
            verbose
    '''

    X = np.array(X_train)
    y = np.array(y_train)

    input_size = X.shape[1]

    model = Sequential([
        Dense(units = units_array[0], input_shape = (input_size,), activation = activations_array[0])
    ])
    
    for i in range(1, hidden_layers_number):
        model.add(Dense(units = units_array[i], activation = activations_array[i]))

    model.add(Dense(units=2, activation='softmax'))

    model.compile(optimizer = tf.keras.optimizers.Adam(learning_rate = learning_rate), loss=loss_function, metrics = metrics_array)    

    model.fit(x=X, y=y, batch_size=10, epochs=30, shuffle=True, verbose=verbose)

    return model
```

<div style="direction:rtl;line-height:300%;text-align:justify;" align="justify"><font face="B nazanin" size=4>
تابع زیر در ورودی, مدل ترین‌شده و داده‌ی تست(یا اعتبارسنجی) را می‌گیرد و در خروجی, دقت مدل را برمی‌گرداند. از خروجی این تابع می‌توانید برای بررسی عملکرد مدل  استفاده کنید.</font></div>

```python
def test_model(model, X_test, y_test, batch_size):
    eval = model.evaluate(x=X_test.to_numpy(), y=y_test.to_numpy(), batch_size=batch_size, verbose= 0)
    return eval[1]
```

<div style="direction:rtl;line-height:300%;text-align:justify;" align="justify"><font face="B nazanin" size=4>
حال با استفاده از توابع بالا کل فرآیند را در ذیل توسعه دهید.
</font></div>

```python
hparams0 = (1, [16], ['relu'], 'sparse_categorical_crossentropy', ['accuracy'], 0.0001, 32, 30, 0)
hparams1 = (2, [16, 32], ['relu', 'relu'], 'sparse_categorical_crossentropy', ['accuracy'], 0.0001, 32, 30, 0)
hparams2 = (2, [16, 16], ['relu', 'relu'], 'sparse_categorical_crossentropy', ['accuracy'], 0.0001, 32, 30, 0)
hparams3 = (2, [16, 32], ['softmax', 'softmax'], 'sparse_categorical_crossentropy', ['accuracy'], 0.0001, 32, 30, 0)
hparams4 = (2, [16, 16], ['softmax', 'softmax'], 'sparse_categorical_crossentropy', ['accuracy'], 0.0001, 32, 30, 0)
hparams5 = (2, [32, 32], ['relu', 'relu'], 'sparse_categorical_crossentropy', ['accuracy'], 0.0001, 32, 30, 0)
hparams6 = (2, [64, 64], ['relu', 'relu'], 'sparse_categorical_crossentropy', ['accuracy'], 0.0001, 32, 30, 0)
hparams7 = (3, [16, 32, 32], ['relu', 'relu', 'relu'], 'sparse_categorical_crossentropy', ['accuracy'], 0.0001, 32, 30, 0)
```

```python
hparams, best_hparam = [hparams0, hparams1, hparams2, hparams3, hparams4, hparams5, hparams6, hparams7], None
val_accs = [] #compare results based on validation set
for param in hparams:
    hidden_layers_number, units_array, activations_array, loss_function, metrics_array, learning_rate, batch_size, epochs ,verbose = param
    model = train_model(X_train, y_train, hidden_layers_number, units_array, activations_array, loss_function, metrics_array, learning_rate, batch_size, epochs ,verbose)
    val_accs.append(test_model(model, X_val, y_val, batch_size))

print(val_accs)
```

```python
best_acc, best_idx = 0, -1
for i in range(len(hparams)):
    if val_accs[i] > best_acc:
        best_acc = val_accs[i]
        best_idx = i

hidden_layers_number, units_array, activations_array, loss_function, metrics_array, learning_rate, batch_size, epochs ,verbose = hparams[best_idx]
best_model = train_model(X_train, y_train, hidden_layers_number, units_array, activations_array, loss_function, metrics_array, learning_rate, batch_size, epochs ,verbose)
test_acc = test_model(best_model, test_X, test_y, batch_size)
print("Model Test Accuracy: {}".format(test_acc))
```
