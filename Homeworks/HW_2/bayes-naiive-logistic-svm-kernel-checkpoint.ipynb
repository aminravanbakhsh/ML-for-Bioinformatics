تابع 
normal_params
به عنوان ورودی یک دیتافریم می‌گیرد که تمام ستون‌های آن پیوسته هستند و در خروجی پارامترهای توزیع نرمال چند متغیری ستون‌ها را برمی‌گرداند
</font></div>

```python
def normal_params(X, y):
    '''
    parameters:
        X: pd dataframe
        y: pd series

    return:
        p0: probability of class 0
        p1: probability of class 1
        m0: mean of class 0
        m1: mean of class 1
        s0: covariance matrix class 0
        s1: covariance matrix class 1
    '''
    

    t0 = y == 0
    t1 = y == 1
    X0 = X[t0].to_numpy()
    X1 = X[t1].to_numpy()
    n0, n1, n = len(X0), len(X1), len(X)
    
    p0 = n0/n
    p1 = n1/n

    m0 = np.sum(X0, axis=0)/n0
    m1 = np.sum(X1, axis=0)/n1

    s0 = np.dot((X0 - m0).T, (X0 - m0)) / n0
    s1 = np.dot((X1 - m1).T, (X1 - m1)) / n1
    
    return p0, p1, m0, m1, s0, s1
```

<div style="direction:rtl;line-height:300%;text-align:justify;" align="justify"><font face="B nazanin" size=4>
تابع زیر با گرفتن یک بردار از ویژگی‌های پیوسته و پارامتر‌های توزیع نرمال چند متغیره، احتمال دیده شدن آن بردار را برمی‌گرداند
</font></div>

```python
def normal_p(x, m, s):
    '''
    parameters:
        x: np.array
            vector of features
        m: np.array
            bayes mean
        s: np.array
            bayes covariance

    return:
        p: int
            probability
    '''

    
    d = m.shape[0]  
    z = np.array(x) - m
    d = m.shape[0]
    det = np.linalg.det(s0)
    arg = np.exp(-0.5 * np.dot(np.dot(z.T, np.linalg.inv(s)), z))
    A =  1 / ((2*np.pi) ** (d) * det) ** 0.5 
    return A * arg
```

```python
def p_x_normal(x, params):
    '''
    parameters:
        x: np.array
            vector of features
        params: tuple of bayes probability
            (p0, p1, m0, m1, s0, s1)

    return:
        p: tuple(int)
            probabilities
    '''
    
    p0, p1, m0, m1, s0, s1 = params
    return p0 * normal_p(x, m0, s0), p1 * normal_p(x, m1, s1)
```

<div style="direction:rtl;line-height:300%;text-align:justify;" align="justify"><font face="B nazanin" size=4>
تابع 
cat_params
با گرفتن یک دیتافریم و اسم یک ستون گسسته از آن احتمال دیده شدن هر یک از مقادیر منحصر به فرد آن ستون را برمی‌گرداند.
</font></div>

```python
def cat_params(df, col):
    '''
    parameters:
        df: pd dataframe
        col: string
            column name
    
    retrun: 
        dic: dictionary of probabilities
    '''

    keys = df[col].unique()
    n = len(df)
    dic = {}
    for key in keys:
        dic[key] = len(df[df[col] == key])/n

    return dic
```

```python
def cat_params_df(df):
    '''
    parameters:
        df: pd dataframe
    
    retrun: 
        dic_arr: dictionary of dictionary of probabilities
    '''

    n = len(df)
    dic_p = {}
    columns = df.columns

    for col in columns:
        dic_p[col] = cat_params(df, col)

    return dic_p
```

```python
def cat_params_dics(df, y):
    '''
    parameters:
        df: pd dataframe
        y: target
    
    retrun: 
        dic0: dictionary of dictionary of probabilities y == 0
        dic1: dictionary of dictionary of probabilities y == 1
    '''

    t0 = y == 0
    t1 = y == 1
    X0 = df[t0]
    X1 = df[t1]
    
    return cat_params_df(X0), cat_params_df(X1)
```

<div style="direction:rtl;line-height:300%;text-align:justify;" align="justify"><font face="B nazanin" size=4>
تابع زیر با گرفتن یک بردار از ویژگی‌های گسسته و یک دیتافریم، احتمال دیده شدن آن بردار را برمی‌گرداند.
</font></div>

```python
def p_x_cat_dic(dic, x):
    
    '''
    parameters:
        dic: dictionary
            probabilities
        x: pd series
            vector of features

    return:
        p: int
            probability
    '''

    columns = dic.keys()
    p = 1
    for col in columns:
        if x[col] in dic[col].keys():
            p *= dic[col][x[col]]
        else:
            p *= 0
            return 0

    return p
```

```python
def p_x_cat(df, x):
    '''
    parameters:
        df: pd dataframe
            discrete dataframe
        x: pd series
            vector of features

    return:
        p: int
            probability
    '''

    columns = df.columns
    p = 1
    for col in columns:
        dic = cat_params(df, col) 
        if x[col] in dic.keys():
            p *= dic[x[col]]
        else:
            p *= 0
            return

    return p
```

```python
def discrete_dics(X, y):
    t0 = y == 0
    t1 = y == 1
    dic_0 = cat_params_df(X[t0])
    dic_1 = cat_params_df(X[t1])
    return dic_0, dic_1
```

<div style="direction:rtl;line-height:300%;text-align:justify;" align="justify"><font face="B nazanin" size=4>
تابع زیر با گرفتن یک دیتافریم، پارامتر‌های مورد نیاز برای توزیع های پیوسته و گسسته و یک بردار از ویژگی‌ها، احتمال دیده شدن آن بردار را در خروجی برمی‌گرداند.
</font></div>

```python
def p_x_multivariate(continuous_params, discrete_dics, discrete_columns, continuous_columns, x):
    
    x_discrete = x[discrete_columns]
    x_continuous = x[continuous_columns]

    dic0, dic1 = discrete_dics

    p_c0, p_c1 = p_x_normal(x_continuous, continuous_params)
    p_d0 = p_x_cat_dic(dic0, x_discrete)
    p_d1 = p_x_cat_dic(dic1, x_discrete)
    

    return p_c0 * p_d0, p_c1 * p_d1
```

```python
def p_df_multivariate(continuous_params, discrete_dics, discrete_columns, continuous_columns, df):
    ps = []
    n = len(df)
    for i in range(n):
        ps.append(p_x_multivariate(continuous_params, discrete_dics, discrete_columns, continuous_columns, df.iloc[i]))
    
    return ps
```

```python
'''
    class for evaluation of target and predict
'''

class Evaluation:

    def __init__(self, predict, target):
        self.predict = np.array(predict)
        self.target = np.array(target)
        correct = np.logical_not(np.logical_xor(self.predict, self.target))
        self.N = len(self.target)
        self.TP = list(np.logical_and(correct, self.target)).count(True)
        self.TN = list(np.logical_and(correct, np.logical_not(self.target))).count(True)
        self.FP = list(np.logical_and(np.logical_not(correct), self.target)).count(True)
        self.FN = list(np.logical_and(np.logical_not(correct), np.logical_not(self.target))).count(True)

    def accuracy(self):
        return (self.TP + self.TN) / self.N

    def precision(self):
        return self.TP/(self.TP + self.FP)

    def recall(self):
        return self.TP/(self.TP + self.FN)

    def sensitivity(self):
        return self.TP/(self.TP + self.FN)
    
    def specificity(self):
        return self.TN/(self.TN + self.FP)

    def f_score(self):
        return 2 / ((1/self.precision()) + (1/ self.recall()))
```

```python
def precision(y_true, y_pred):
    eval = Evaluation(y_pred, y_true)
    return eval.precision()
```

```python
def recall(y_true, y_pred):
    eval = Evaluation(y_pred, y_true)
    return eval.recall()
```

```python
def f_score(y_true, y_pred):
    eval = Evaluation(y_pred, y_true)
    return eval.f_score()
```

تابع زیر یک عدد میگیرد و با توجه به پارامتر های توزیع نرمال مربوط به آن احتمال دیده شدن آن را برمی‌گرداند
</font></div>

```python
def p_x_single_normal(x, m, s):
    '''
    parameters:
        x: np.array
            vector of features
        m: np.array
            bayes mean
        s: np.array
            bayes covariance

    return:
        p: int
            probability
    '''

    z = x - m
    det = s
    arg = np.exp(-0.5 * z * z / s)
    A =  1 / ((2*np.pi) * det) ** 0.5 
    return A * arg
```

```python
def p_single_normal(x, params):
    p0, p1, m0, m1, s0, s1 = params
    return p0 * p_x_single_normal(x, m0, s0), p1 * p_x_single_normal(x, m1, s1)
```

<div style="direction:rtl;line-height:300%;text-align:justify;" align="justify"><font face="B nazanin" size=4>
تابع زیر مشابه قبل ولی اینبار با فرض استقلال تمام ویژگی‌ها با گرفتن یک دیتافریم، پارامتر‌های مورد نیاز برای توزیع های پیوسته و گسسته و یک بردار از ویژگی‌ها، احتمال دیده شدن آن بردار را در خروجی برمی‌گرداند.
</font></div>

```python
def p_x_dic_normal(X, y):
    '''
    parameters:
        X: countinuous dataframes
        y: target

    return:
        dic: dictionray
            normal distributions
    '''
    dic = {}
    columns = X.columns
    for col in columns:
        dic[col] = single_normal_param(X, col, y)
    
    return dic
```

```python
def p_x_cat_dic_continuous(dic, x):
    
    '''
    parameters:
        dic: dictionary
            probabilities
        x: pd series
            vector of features

    return:
        p: int
            probability
    '''

    columns = dic.keys()
    p0 = 1
    p1 = 1
    for col in columns:
        p_col0, p_col1 = p_single_normal(x[col], dic[col])
        p0 *= p_col0
        p1 *= p_col1

    return p0, p1
```

```python
def p_x_single(continuous_dics, discrete_dics, discrete_columns, continuous_columns, x):
    x_discrete = x[discrete_columns]
    x_continuous = x[continuous_columns]

    dic0, dic1 = discrete_dics

    p_c0, p_c1 = p_x_cat_dic_continuous(continuous_dics, x)
    p_d0 = p_x_cat_dic(dic0, x_discrete)
    p_d1 = p_x_cat_dic(dic1, x_discrete)
    

    return p_c0 * p_d0, p_c1 * p_d1
```

```python
def p_df_single(continuous_params, discrete_dics, discrete_columns, continuous_columns, df):
    ps = []
    n = len(df)
    for i in range(n):
        ps.append(p_x_single(continuous_params, discrete_dics, discrete_columns, continuous_columns, df.iloc[i]))
    
    return ps
```

```python
def select_class(p0, p1):
    if p0 > p1:
        return 0
    else:
        return 1
```

```python
def select_class_array(ps):
    classes = []
    for p in ps:
        classes.append(select_class(p[0], p[1]))

    return classes
```

```python
def select_class_ratio(p0, p1, ratio):
    if p0 > p1 * ratio:
        return 0
    else:
        return 1
```

```python
def select_class_array_ratio(ps, ratio):
    classes = []
    for p in ps:
        classes.append(select_class_ratio(p[0], p[1], ratio))

    return classes
```

```python
class Logistic_Regression:

    def __init__(self, X, y):
        self.X = np.array(X)
        self.y = np.array(y).reshape((self.X.shape[0], 1))
        self.w = np.zeros((1, self.X.shape[1]))
        self.const = np.ones((self.X.shape[0], 1))
        self.b = 0

    def gradiant(self, landa):
        '''
        parameters:
            X: np.array (n, d)
                dataframe features
            y: np.array (n, 1)
                targets
            w: np.array (1, d)
                weights
            landa: int
                regularization ratio

        return gradiant
        '''
        X = self.X
        y = self.y
        w = self.w
        b = self.b
        const = self.const

        z = np.dot(X, w.T) + b
        f = 1 / (1 + np.exp(-z))
        z = f - y

        gw = np.dot(z.T, X) + landa * w
        gb = np.sum(z) + landa * b

        return (gw, gb)

    def loss(self, landa):
        X = self.X
        y = self.y
        w = self.w
        b = self.b
        const = self.const

        z = np.dot(X, w.T) + b
        f = 1 / (1 + np.exp(-z))
        loss = -np.dot(y.T, np.log(f)) - np.dot((1 - y).T, np.log(1- f)) + landa * (np.dot(w, w.T) + b*b)
        return loss

    def predict(self, X):
        x = np.array(X)
        f = np.dot(x, self.w.T) + self.b
        z = 1 / (1 + np.exp(f))
        z = z.reshape((z.shape[0],))
        return (z >= 0.5).astype(int)

    def fit(self, landa, lr, epochs):
        for _ in range(epochs):
            gw, gb = self.gradiant(landa)
            self.w -= lr * gw
            self.b -= lr* gb
```
