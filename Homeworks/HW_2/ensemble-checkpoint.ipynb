```python
'''
    class for evaluation of target and predict
'''

class Evaluation:

    def __init__(self, predict, target):
        self.predict = np.array(predict)
        self.target = np.array(target)
        correct = np.logical_not(np.logical_xor(self.predict, self.target))
        self.N = len(self.target)
        self.TP = list(np.logical_and(correct, self.target)).count(True)
        self.TN = list(np.logical_and(correct, np.logical_not(self.target))).count(True)
        self.FP = list(np.logical_and(np.logical_not(correct), self.target)).count(True)
        self.FN = list(np.logical_and(np.logical_not(correct), np.logical_not(self.target))).count(True)

    def accuracy(self):
        return (self.TP + self.TN) / self.N

    def precision(self):
        return self.TP/(self.TP + self.FP)

    def recall(self):
        return self.TP/(self.TP + self.FN)

    def sensitivity(self):
        return self.TP/(self.TP + self.FN)
    
    def specificity(self):
        return self.TN/(self.TN + self.FP)

    def f_score(self):
        return 2 / ((1/self.precision()) + (1/ self.recall()))
```

```python
class RandomForest:

    def __init__(self, X, y):
        self.X = X
        self.y = y
        self.trees = []

    def bootstrap_sampling(self):
        n = len(self.X)
        idx = np.random.choice(np.arange(n), replace=True, size=n)
        Xn = self.X.iloc[idx]
        yn = self.y.iloc[idx]
        return Xn, yn

    def fit(self, trees_number):
        trees = []
        for _ in range(trees_number): 
            X, y = self.bootstrap_sampling()
            tree = self.train_tree(X, y)
            trees.append(tree)

        self.trees = trees

    def map(self, array):
        return 2 * array - 1

    def demap(self, array):
        return (1 + array)/2

    def predict(self, X_test):
        n = len(X_test)
        score = np.zeros((n,))
        trees_number = len(self.trees)

        for i in range(trees_number):
            score += self.map(self.trees[i].predict(X_test))

        return self.demap(np.sign(score))


    def train_tree(self, X, y):
        DTC = DecisionTreeClassifier(max_features='sqrt', criterion='entropy')
        DTC.fit(X, y)
        return DTC
```

```python
class Adaboost:

    def __init__(self,X,y):
        self.X = X
        self.y = y
        self.trees = []
        self.alphas = []
        self.losses = []

    def fit(self, epochs):
        trees = []
        alphas = []        

        n = len(self.X)
        w = np.ones((n,))
        scores = np.zeros((n,))

        for _ in range(epochs):
            DTC = DecisionTreeClassifier(max_depth=1, criterion='entropy')
            DTC.fit(self.X, self.y, sample_weight=w)
            
            predict = DTC.predict(self.X)           

            nI = np.logical_xor(predict, self.y)
            I = np.logical_not(nI)
            epsilon = np.sum(nI * w) / np.sum(w)

            if epsilon >= 10 ** -7 and epsilon <= 1 - 10 ** -7:
                alpha = 0.5 * np.log((1 - epsilon)/epsilon)
                w = w * np.exp(-alpha * I)
                alphas.append(alpha)
                trees.append(DTC)
            else:
                break

        self.trees = trees
        self.alphas = alphas

    def map(self, array):
        return 2 * array - 1

    def demap(self, array):
        return (1 + array)/2

    def predict(self, X_test):
        n = len(X_test)
        score = np.zeros((n,))
        trees_number = len(self.trees)

        for i in range(trees_number):
            score += self.alphas[i] * self.map(self.trees[i].predict(X_test))

        return self.demap(np.sign(score))

    def estimate(self, X_test, y_test):
        n = len(X_test)
        score = np.zeros((n,))
        trees_number = len(self.trees)
        losses = []

        for i in range(trees_number):
            score += self.alphas[i] * self.map(self.trees[i].predict(X_test))
            predict = self.demap(np.sign(score))
            losses.append(self.loss(predict, y_test))

        return losses

    def loss(self, predict, target):
        return np.sum((predict != target).astype(int))
```

```python
X_train, X_val, y_train, y_val = sklearn.model_selection.train_test_split(train_X, train_y, test_size = 0.2)
```

```python
tree_numbers = [20, 50, 80, 100, 150, 200]
test_loss = []
val_loss = []
train_loss = []
for number in tree_numbers:
    RF = RandomForest(X_train, y_train)
    RF.fit(number)

    predict_train = RF.predict(X_train)
    eval_train = Evaluation(predict_train, y_train)
    train_loss.append(1 - eval_train.accuracy())

    predict_val = RF.predict(X_val)
    eval_val = Evaluation(predict_val, y_val)
    val_loss.append(1 - eval_val.accuracy())

    predict_test = RF.predict(test_X)
    eval_test = Evaluation(predict_test, test_y)
    test_loss.append(1 - eval_test.accuracy())
```
