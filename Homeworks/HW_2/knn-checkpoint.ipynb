```python
def knn_decide(v, df, y, n):
    nearest_idx = nearest(v, df, n)
    ys = y[nearest_idx]
    if np.sum(ys) >= n/2:
        return 1
    else:
        return 0
```

```python
class KNN:

    def __init__(self, X, y):
        self.X = X
        self.y = y

    def distance(self, v):
        x = self.X - v
        z = np.sum(x**2, axis=1)
        return z

    def nearest(self, v, n):
        d = self.distance(v)
        return d.sort_values().index[:n]

    def v_decide(self, v, n):
        nearest_idx = self.nearest(v, n)
        ys = self.y[nearest_idx]
        if np.sum(ys) >= n/2:
            return 1
        else:
            return 0

    def df_decide(self, X, n):
        predict = []
        size = len(X)
        for i in range(size):
            predict.append(self.v_decide(X.iloc[i], n))
        
        return predict
```

```python
'''
    class for evaluation of target and predict
'''

class Evaluation:

    def __init__(self, predict, target):
        self.predict = np.array(predict)
        self.target = np.array(target)
        correct = np.logical_not(np.logical_xor(self.predict, self.target))
        self.N = len(self.target)
        self.TP = list(np.logical_and(correct, self.target)).count(True)
        self.TN = list(np.logical_and(correct, np.logical_not(self.target))).count(True)
        self.FP = list(np.logical_and(np.logical_not(correct), self.target)).count(True)
        self.FN = list(np.logical_and(np.logical_not(correct), np.logical_not(self.target))).count(True)

    def accuracy(self):
        return (self.TP + self.TN) / self.N

    def precision(self):
        return self.TP/(self.TP + self.FP)

    def recall(self):
        return self.TP/(self.TP + self.FN)

    def sensitivity(self):
        return self.TP/(self.TP + self.FN)
    
    def specificity(self):
        return self.TN/(self.TN + self.FP)

    def f_score(self):
        return 2 / ((1/self.precision()) + (1/ self.recall()))
```

```python
X_train, X_val, y_train, y_val = sklearn.model_selection.train_test_split(train_X, train_y, test_size=0.2, shuffle=True)
```

### k = 1

```python
knn1 = KNN(X_train, y_train)
predict = knn1.df_decide(X_val, 1)
eval = Evaluation(predict, y_val)
print('n: 1')
print('accuracy:', eval.accuracy())
print('f_score:', eval.f_score())
print('precision:', eval.precision())
print('recall:', eval.recall())
```
